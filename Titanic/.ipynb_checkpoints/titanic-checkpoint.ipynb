```python
import sklearn.preprocessing as preprocessing
def preprocess_data(dt):
    ## 处理年龄
    avg = dt.Age.mean(skipna=True)
    dt.Age=dt.Age.fillna(avg)
#     dt[dt.Age.isna()]
    
    favg = dt.Fare.mean(skipna=True)
    dt.Fare =dt.Fare.fillna(favg)
    
    ## 处理性别
    def sex_level(x):
        if x =='female':
            return 0
        else:
            return 1
    dt['Sex_l']= dt.Sex.apply(sex_level)
    
    ## 
    def check_nan(x):
        if x != x:
            return 1
        else:
            return 0

    dt['Cabin_l'] = dt.Cabin.apply(check_nan)

    scaler = preprocessing.StandardScaler()
    # age_filter_param = scaler.fit()
    dt['Age_scaled'] = preprocessing.scale(dt['Age'])
    dt['Fare_scaled']= preprocessing.scale(dt['Fare'])
    dt.Embarked = dt.Embarked.fillna('N')
    dt['Embarked_l'] = dt.Embarked.apply(lambda x: ord(x)-ord('C'))
    train_df = dt.filter(items=['Survived', 'Pclass','Sex_l','Age_scaled','SibSp','Parch','Fare_scaled','Cabin_l','Embarked_l'])
#     print(train_df.head())
    return train_df.as_matrix()
    
    
```

```python
t1 = pd.read_csv('data/train.csv',sep=',')
preprocess_data(t1)
```

```python
from sklearn import linear_model
def train_process(df):
    y = df[:,0]
    x = df[:,1:]
    clf = linear_model.LogisticRegression(C=1.0,penalty='l1',tol=1e-5)
    clf.fit(x,y)
    return clf
```

```python
clf = train_process(preprocess_data(t1))
```

```python
t2 = pd.read_csv('data/test.csv',sep=',')
t2[t2.Fare.isna()]
```

```python


d2 = preprocess_data(t2)
pd2 = clf.predict(d2)
```

```python
results = pd.DataFrame({'PassengerId':t2['PassengerId'],'Survived':pd2.astype(np.int32)})
```

```python
results.to_csv('predict.csv',index=False)
```
